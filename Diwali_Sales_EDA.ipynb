# Exploratory Data Analysis

### Gender

```python
# male vs. female count
chart_gender = sns.countplot(x='Gender',data=df)

for bars in chart_gender.containers:
    chart_gender.bar_label(bars)

sns.set_theme(rc={'figure.figsize':(5,5)})
```

```python
# purchasing power -- male vs. female
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gender)

sns.set_theme(rc={'figure.figsize':(5,5)})
```

From the graphs above, we observe that majority of the buyers are females and they spend more than males  

### Age

```python
# age group buyer count -- male vs. female
chart_age_group = sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in chart_age_group.containers:
    chart_age_group.bar_label(bars)

sns.set_theme(rc={'figure.figsize':(15,5)})
```

```python
# purchasing power -- age group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

From the graphs above, we observe that majority of the buyers are 26-35 years old with the highest purchasing capacity  

### States

```python
# States buyer count
sales_states = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set_theme(rc={'figure.figsize':(18,5)})

sns.barplot(x='State', y='Orders', data=sales_states)
```

```python
# States purchasing power
sales_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set_theme(rc={'figure.figsize':(18,5)})

sns.barplot(x='State', y='Amount', data=sales_states)
```

From the graphs above, we observe that most of the buyers hail from the states of Uttar Pradesh, Maharashtra, and Karnataka with the highest purchasing power

### Marital Status

```python
# Married (0) vs. single (1) buyer count
chart_marital_status = sns.countplot(x='Marital_Status',data=df)

for bars in chart_marital_status.containers:
    chart_marital_status.bar_label(bars)

sns.set_theme(rc={'figure.figsize':(7,5)})
```

```python
# Married (0) vs. single (1) purchasing power gender-wise
sales_marital_status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Marital_Status', y='Amount', data=sales_marital_status, hue='Gender')

sns.set_theme(rc={'figure.figsize':(10,5)})
```

From the graphs above, we observe that majority of the buyers are married, and maximum purchase is made by married females

### Occupation

```python
# Occupation buyer count
chart_occupation = sns.countplot(x='Occupation', data=df)

for bars in chart_occupation.containers:
    chart_occupation.bar_label(bars)

sns.set_theme({'figure.figsize':(20,5)})
```

```python
# Purchasing power occupation-wise
sales_occupation = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Occupation', y='Amount', data=sales_occupation)

sns.set_theme(rc={'figure.figsize':(20,5)})
```

From the graphs above, we observe that most of the buyers work in the IT, Healthcare, and Aviation sectors with the highest purchasing capacity

### Product Category

```python
# Products sold by category -- count
chart_product = sns.countplot(x='Product_Category',data=df)

for bars in chart_product.containers:
    chart_product.bar_label(bars)

sns.set_theme(rc={'figure.figsize':(25,10)})
```

```python
# Products sold by category -- amount
sales_product = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.barplot(x='Product_Category', y='Amount', data=sales_product)

sns.set_theme(rc={'figure.figsize':(20,5)})
```

From the graphs above, we observe that clothing items were bought the most, but maximum revenue was generated from food products
